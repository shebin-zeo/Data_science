```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
data=pd.read_csv(r"C:\Users\shebi\Downloads\Salary_Data.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x=data[["YearsExperience"]]
y=data["Salary"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
predict=knn.predict(x_test)
predict
```

```python
def Salary_Knn():
    try:
        input_data=float(input("Enter the year of experience : "))
        years=[[input_data]]
        prediction=knn.predict(years)
        print(f"The Salary with {input_data} : " ,prediction[0])
    except ValueError:
        print("Enter valid year")
Salary_Knn()
```
